# Implement a linear regression model to predict the prices of houses based on their square footage and the number of bedrooms and bathrooms.

Preprocessing

```python
# find the columns with null values
null_counts = housing_train.isna().sum()

print("Null value counts per column:")
for column, count in null_counts.items():
    print(f"{column}: {count}")
```

```python
# remove 'SalePrice' column from housing_train and store it in a separate variable
target_variable = housing_train.pop('SalePrice')

# drop rows with any NaN values from housing_train
housing_train = housing_train.dropna(axis = 1)
housing_test = housing_test.dropna(axis = 1)

housing_train_columns = housing_train.columns
housing_test_columns = housing_test.columns

# find the intersection of columns between housing_train and housing_test
common_columns = housing_train.columns.intersection(housing_test_columns)
common_columns
```

```python
housing_train = housing_train[common_columns]
housing_train
```

```python
housing_train.isna().sum()
```

```python
housing_test = housing_test[common_columns]
housing_test
```

```python
# deal with cathegorical columns
categorical_columns = housing_train.select_dtypes(include=['object']).columns.tolist()

print("Categorical columns:")
for col in categorical_columns:
    print(col)
```

```python
# work with OrdinalEncoder
from sklearn.preprocessing import OrdinalEncoder

encoder = OrdinalEncoder()
housing_train[categorical_columns] = encoder.fit_transform(housing_train[categorical_columns])
housing_train[categorical_columns]
```

```python
housing_test[categorical_columns] = encoder.fit_transform(housing_test[categorical_columns])
housing_test[categorical_columns]
```

```python
# Standardize the non-cathegorical values
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
categorical_columns.append('Id')
columns_to_scale = [col for col in housing_train.columns if col not in categorical_columns]
housing_train[columns_to_scale] = scaler.fit_transform(housing_train[columns_to_scale])
housing_train
```

```python
housing_test[columns_to_scale] = scaler.fit_transform(housing_test[columns_to_scale])
housing_test
```

Training and predicting

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
# split the data into features and labels
X_train = housing_train
y_train = target_variable

# train
linear_reg = LinearRegression()
linear_reg.fit(X_train, y_train)
```

```python
# predict
y_pred = linear_reg.predict(housing_test)
```

```python
# add the predicted saleprice to dataset
housing_test['SalePrice'] = y_pred
housing_test
```
